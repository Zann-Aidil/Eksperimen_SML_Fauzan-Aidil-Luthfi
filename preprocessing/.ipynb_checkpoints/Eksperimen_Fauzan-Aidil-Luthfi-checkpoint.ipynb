# **1. Perkenalan Dataset**

Dataset yang digunakan adalah **Titanic - Machine Learning from Disaster** dari Kaggle.

# **5. Data Preprocessing**

```python
# ============================================================
# STEP 1: Drop kolom yang tidak relevan
# ============================================================
df_processed = df.copy()

# PassengerId, Name, Ticket tidak berguna untuk model
# Cabin terlalu banyak missing (77%), lebih baik buat fitur baru
df_processed['Has_Cabin'] = df_processed['Cabin'].notnull().astype(int)
df_processed.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)

print(f'Setelah drop kolom — Shape: {df_processed.shape}')
print(f'Kolom tersisa: {list(df_processed.columns)}')
```

```python
# ============================================================
# STEP 2: Handle Missing Values
# ============================================================

# Age — isi dengan median (lebih robust terhadap outlier)
df_processed['Age'].fillna(df_processed['Age'].median(), inplace=True)

# Embarked — isi dengan modus (kategori paling sering)
df_processed['Embarked'].fillna(df_processed['Embarked'].mode()[0], inplace=True)

print('Missing values setelah dihandle:')
print(df_processed.isnull().sum())
print('Handle missing values selesai')
```

```python
# ============================================================
# STEP 3: Feature Engineering
# ============================================================

# FamilySize = SibSp + Parch + 1 (dirinya sendiri)
df_processed['FamilySize'] = df_processed['SibSp'] + df_processed['Parch'] + 1

# IsAlone = 1 jika naik sendirian
df_processed['IsAlone'] = (df_processed['FamilySize'] == 1).astype(int)

# AgeGroup — kategori umur
df_processed['AgeGroup'] = pd.cut(
    df_processed['Age'],
    bins=[0, 12, 18, 35, 60, 100],
    labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior']
)

print('Fitur baru berhasil dibuat')
print(df_processed[['FamilySize', 'IsAlone', 'AgeGroup']].head())
```

```python
# ============================================================
# STEP 4: Encoding Kategorikal
# ============================================================

# Sex — Label Encoding (male=1, female=0)
le = LabelEncoder()
df_processed['Sex'] = le.fit_transform(df_processed['Sex'])

# Embarked — One-Hot Encoding
df_processed = pd.get_dummies(df_processed, columns=['Embarked'], drop_first=True)

# AgeGroup — One-Hot Encoding
df_processed = pd.get_dummies(df_processed, columns=['AgeGroup'], drop_first=True)

print('Encoding selesai')
print(f'Shape setelah encoding: {df_processed.shape}')
print(f'Kolom: {list(df_processed.columns)}')
```

```python
# ============================================================
# STEP 5: Deteksi & Handle Outlier (IQR Method)
# ============================================================

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before = len(df)
    df = df[(df[column] >= lower) & (df[column] <= upper)]
    after = len(df)
    print(f'  {column}: hapus {before - after} outlier')
    return df

print('Hapus outlier:')
df_processed = remove_outliers_iqr(df_processed, 'Fare')
df_processed = remove_outliers_iqr(df_processed, 'FamilySize')

print(f'Shape setelah remove outlier: {df_processed.shape}')
```

```python
# ============================================================
# STEP 6: Normalisasi Fitur Numerik
# ============================================================

scaler = StandardScaler()
cols_to_scale = ['Age', 'Fare', 'FamilySize']

df_processed[cols_to_scale] = scaler.fit_transform(df_processed[cols_to_scale])

print('Normalisasi selesai')
print(df_processed[cols_to_scale].describe().round(2))
```

```python
# ============================================================
# STEP 7: Simpan Hasil Preprocessing
# ============================================================
import os

os.makedirs('../titanic_preprocessing', exist_ok=True)
df_processed.to_csv('../titanic_preprocessing/train_preprocessed.csv', index=False)

print(f'Dataset preprocessed disimpan — Shape final: {df_processed.shape}')
print('Preprocessing SELESAI — siap dilatih!')
df_processed.head()
```
